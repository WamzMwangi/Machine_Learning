## 1. Setting the Background

Floods are devastating natural events that result in significant damage to property, disrupt economies, and sadly, loss of life. It is very crucial for emergency response teams and governments alike to be well prepared to handle the negative impacts of floods.
Flood prediction helps governments and emergency teams identify areas most vulnerable to flooding and implement the necessary proactive measures such as planning evacuation routes, early warning systems,erecting dykes to control swelling rivers etc..

In this project, I will use machine learning models to predict the likelihood floods occuring using rainfall data from several districts and states in Malaysia, between the year 2000 and 2010.
